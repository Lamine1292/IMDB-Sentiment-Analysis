##   **Introduction**

In this notebook, we perform a complete Exploratory Data Analysis (EDA) and visualization of two classic movie review datasets:

### 1. Rotten Tomatoes Movie Reviews (RT Polarity Dataset)

* 10,662 short review sentences (5,331 positive, 5,331 negative).
* Pre-tokenized, clean, and balanced.
* Source: Pang & Lee (2004).


### 2. Cornell Movie Review Data (Polarity Dataset v2.0)

* 2,000 full-length reviews (1,000 pos, 1,000 neg).
* Raw text with HTML remnants, punctuation, and narrative structure.
* Source: Cornell University.



### We will:

* Load and inspect both datasets

* Clean and preprocess the text

* Compare their characteristics

* Visualize word frequencies, lengths, n-grams, and sentiment

* Build word clouds and statistical summaries
